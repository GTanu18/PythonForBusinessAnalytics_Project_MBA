## Task 3: Data Analysis

-  ### Subtask 3.1: Convert units of gross and budget to millions for all movies in the dataset. (2 marks)

```python
movies["gross"] = movies[["gross"]].apply(lambda x:x/1000000)
movies["budget"] = movies[["budget"]].apply(lambda x:x/1000000)
# We have used lambda functions to convert values of each row in columns "gross" and "budget" to reflect in "$ Million" from $
movies
```

-  ### Subtask 3.2:  Display the 10 movies with the highest profit. (Blockbuster Movies playlist) (2 marks)

```python
movies["profit"]=movies["gross"]-movies["budget"]
# We added a column named "profit", with values profit=gross-budget
Blockbuster_movies=movies.sort_values(["profit"],ascending=False).head(10)
# We sorted the values of dataset in descending order based on profit column and extracted only the top 10 rows.
Blockbuster_movies
```

-  ### Subtask 3.3: Display the top 10 movies (based on IMDb ratings) with a minimum of 25,000 ratings (Most Loved movies playlist) (4 marks)

```python
most_loved_movies=movies[movies["num_voted_users"]>=25000]
#We have filtered about the dataset of movies having ratings more than or equal to 25000.
Top_10_movies=most_loved_movies.sort_values(["imdb_score"], ascending=False).head(10)
# We sorted the values of dataset in descending order based on IMDb sccore column and extracted only the top 10 rows which will give us top 10 most loved movies.
Top_10_movies
```

- ### Subtask 3.4: Display the names of  the top 5 directors based on IMDb ratings with a minimum of 5 movies (Top Directors Playlist)  (4 marks)

```python

temp_movies = movies
#We have created a replica of movies dataset for performing a temporary operation that would change the actual data of dataset.

temp_movies["movie_title"]=1
# When creating a pivot table with respect to a column as an index, the resultant dataframe only have numerial data and the columns with non-numeric data types have dropped.
# So to be able to count each movie, we have assigned value 1 to each movie.

pivotted_dir_name = movies.pivot_table(index=["director_name"], aggfunc={'movie_title':np.sum, 'imdb_score':np.mean})
#Created a pivot table and applied aggregate function of mean to imdb score and sum to movie title column.

top5_dir_name = pivotted_dir_name[pivotted_dir_name["movie_title"]>=5].sort_values("imdb_score",ascending=False).head(5)

# In the above pivot table we got avg. imdb_scores for each director for all of their movies,but here we filtering the dataset for extracting the above data for directors with more than equal to 5 movies.

# We sorted the values of dataset in descending order based on IMDb sccore column.

top5_dir_name
```

 Subtask 3.5.1: Find the most popular genre of movies (by highest average rating) (4 marks)

```python
genres_split=movies["genres"].str.split("|",expand=True)
# we are separating multiple genres by using delimeter.

movies["Genre1"]=genres_split[0]
groupedby_genres=movies.groupby(["Genre1"])
#We used groupby to group the movies as per genres.

most_Popular_genres=groupedby_genres["imdb_score"].mean().sort_values(ascending=False).head(1)
#We have caluated the mean of imdb scores to find out ratings and then sort based on the rating and taking out the most popular genre as per highest rating.
most_Popular_genres
```

-  ### Subtask 3.5.2: Find the most popular genre of movies (by total gross) (4 marks)

```python
most_Popular_genres=groupedby_genres["gross"].sum().sort_values(ascending=False).head(1)

#We have caluated the sum of gross column and then sort based on it and taking out the most popular genre as per highest total gross value.


most_Popular_genres
```
